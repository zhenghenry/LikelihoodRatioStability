# Define the problem and optimizer to analyze the metrics for

```python
'''
The files with the metrics are stored in 'prob_type/optimizer_type/metrics/*' where prob_type can be 'ALEPH, Gaussian, NF'. The name of the file
will be *.npy and the name will be the value of the weighted_error. 
'''
prob_type = 'NF'  # Choose from ALEPH, Gaussian , NF
optimizer = 'ECD_q1' # Choose from ECD_q1, ECD_q1_F2, adam, lion

def load_mae_from_file(prob_type, results):
    if prob_type == 'mlc':
        linear_maes = []
        square_maes = []
        exponl_maes = []

        for result in results:
            if result['classifier'] == 'linear':
                linear_maes.append(result['mae'])
            elif result['classifier'] == 'square':
                square_maes.append(result['mae'])
            elif result['classifier'] == 'exponl':
                exponl_maes.append(result['mae'])
        mae_dict = dict(linear = np.array(linear_maes), square = np.array(square_maes), exponl = np.array(exponl_maes))
    elif prob_type == 'bce':
        odds_maes = []
        probit_maes = []
        arctan_maes = []

        for result in results:
            if result['classifier'] == 'odds':
                odds_maes.append(result['mae'])
            elif result['classifier'] == 'probit':
                probit_maes.append(result['mae'])
            elif result['classifier'] == 'arctan':
                arctan_maes.append(result['mae'])
        mae_dict = dict(odds = np.array(odds_maes), probit = np.array(probit_maes), arctan = np.array(arctan_maes))
    return mae_dict
        
```

```python
def get_mae_std(mae_dict):
    for k, v in mae_dict.items():
        filt_v = v[(v<1) & (v == v)]
        var_val = np.var(filt_v)
        mean_val = np.mean(filt_v)
        print(f'Classifier: {k} MAE: {mean_val} Var: {var_val}\n')
```

## Looking at a slice on physics dimension

```python
from nflows.flows.base import Flow
from nflows.distributions.normal import StandardNormal
from nflows.transforms.base import CompositeTransform
from nflows.transforms.autoregressive import MaskedAffineAutoregressiveTransform
from nflows.transforms.permutations import ReversePermutation
```

```python
d=6
num_layers = 5
base_dist = StandardNormal(shape=[d])

transforms = []
for _ in range(num_layers):
    transforms.append(ReversePermutation(features=d))
    transforms.append(MaskedAffineAutoregressiveTransform(features=d, 
                                                          hidden_features=8))
transform = CompositeTransform(transforms)

dat_target = Flow(transform, base_dist).to('cpu')
ckpt = torch.load('flows/nflows/dat/8/best/ckpt_67',map_location=torch.device("cpu"))
dat_target.load_state_dict(ckpt['model_state_dict'])
dat_target.eval()


num_layers = 5
base_dist = StandardNormal(shape=[d])

transforms = []
for _ in range(num_layers):
    transforms.append(ReversePermutation(features=d))
    transforms.append(MaskedAffineAutoregressiveTransform(features=d, 
                                                          hidden_features=8))
transform = CompositeTransform(transforms)

sim_target = Flow(transform, base_dist).to('cpu')
ckpt = torch.load('flows/nflows/sim/8/best/ckpt_39',map_location=torch.device("cpu"))
sim_target.load_state_dict(ckpt['model_state_dict'])
sim_target.eval();
lr = lambda x: torch.exp(dat_target.log_prob(x) - sim_target.log_prob(x))
```

```python
def odds_lr(model, m = 0, s = 1):
    def model_lr(x):
        f = model((x - m) / s)
        return torch.squeeze(f / (1. - f))
    return model_lr
def probit_lr(model, m = 0, s = 1):
    def model_lr(x):
        f = model((x - m) / s)
        return torch.squeeze(probit(f) / (1. - probit(f)))
    return model_lr
def arctan_lr(model, m = 0, s = 1):
    def model_lr(x):
        f = model((x - m) / s)
        return torch.squeeze(t_arctan(f) / (1. - t_arctan(f)))
    return model_lr    
```

```python
X_trn = torch.from_numpy(X)
y_trn = torch.from_numpy(y)

median = np.median(X_trn[y_trn == 1], axis = 0)
median_m = median[0]
median_mults = median[1]
median_w = median[2]
median_sdjm = median[3]
median_sjr = median[4]
median_pT = median[5]

def get_slice(index_of_slice):
    if index_of_slice == 0:
        # Get a one-dimensional slice of test points.
        ms = np.linspace(0, 170, 1201).reshape(-1, 1)
        mults = np.repeat(median_mults, len(ms)).reshape(-1, 1)
        ws = np.repeat(median_w, len(ms)).reshape(-1, 1)
        sdjms = np.repeat(median_sdjm, len(ms)).reshape(-1, 1)
        sjrs = np.repeat(median_sjr, len(ms)).reshape(-1, 1)
        pTs = np.repeat(median_pT, len(ms)).reshape(-1, 1)
        plt_var = ms
    elif index_of_slice == 1:
        # Get a one-dimensional slice of test points.
        mults = np.linspace(1,80,1201).reshape(-1, 1)
        ms = np.repeat(median_m, len(mults)).reshape(-1, 1)
        sdjms = np.repeat(median_sdjm, len(mults)).reshape(-1, 1)
        ws = np.repeat(median_w, len(ms)).reshape(-1, 1)
        sjrs = np.repeat(median_sjr, len(mults)).reshape(-1, 1)
        pTs = np.repeat(median_pT, len(mults)).reshape(-1, 1)     
        plt_var = mults   
    elif index_of_slice == 2:
        # Get a one-dimensional slice of test points.
        ws = np.linspace(0,0.7,1201).reshape(-1,1)
        mults = np.repeat(median_mults, len(ws)).reshape(-1, 1)
        ms = np.repeat(median_m, len(ws)).reshape(-1, 1)
        sdjms = np.repeat(median_sdjm, len(ws)).reshape(-1, 1)
        sjrs = np.repeat(median_sjr, len(ws)).reshape(-1, 1)
        pTs = np.repeat(median_pT, len(ws)).reshape(-1, 1) 
        plt_var = ws
    elif index_of_slice == 3:
        # Get a one-dimensional slice of test points.
        sdjms = np.linspace(-18,-2,1201).reshape(-1,1)
        mults = np.repeat(median_mults, len(sdjms)).reshape(-1, 1)
        ms = np.repeat(median_m, len(sdjms)).reshape(-1, 1)
        ws = np.repeat(median_w, len(sdjms)).reshape(-1, 1)
        sjrs = np.repeat(median_sjr, len(sdjms)).reshape(-1, 1)
        pTs = np.repeat(median_pT, len(sdjms)).reshape(-1, 1) 
        plt_var = sdjms
    elif index_of_slice == 4:
        # Get a one-dimensional slice of test points.
        sjrs = np.linspace(0,1.4,1201).reshape(-1,1)
        mults = np.repeat(median_mults, len(sjrs)).reshape(-1, 1)
        ms = np.repeat(median_m, len(sjrs)).reshape(-1, 1)
        ws = np.repeat(median_w, len(sjrs)).reshape(-1, 1)
        sdjms = np.repeat(median_sdjm, len(sjrs)).reshape(-1, 1)
        pTs = np.repeat(median_pT, len(sjrs)).reshape(-1, 1) 
        plt_var = sjrs
    elif index_of_slice == 5:
        # Get a one-dimensional slice of test points.
        pTs = np.linspace(10,1700,1201).reshape(-1,1)
        mults = np.repeat(median_mults, len(pTs)).reshape(-1, 1)
        ms = np.repeat(median_m, len(pTs)).reshape(-1, 1)
        ws = np.repeat(median_w, len(pTs)).reshape(-1, 1)
        sdjms = np.repeat(median_sdjm, len(pTs)).reshape(-1, 1)
        sjrs = np.repeat(median_sjr, len(pTs)).reshape(-1, 1) 
        plt_var = pTs
    xs = np.hstack([ms, mults, ws, sdjms, sjrs, pTs])
    xs = torch.tensor(xs, dtype = torch.float32)
    return xs, plt_var
def get_lower_upper(slice_index):
    xs, plt_var = get_slice(slice_index)
    n = plt.hist(X_mae[y_mae == 1][:,slice_index], bins = plt_var.reshape(-1)[::10], density=True)
    lower, upper=(0.5*(n[1][0:-1]+n[1][1:]))[n[0] > n[0].max()/2][0],(0.5*(n[1][0:-1]+n[1][1:]))[n[0] > n[0].max()/2][-1]
    plt.close()
    print(lower, upper)
    return lower, upper
```

```python
with open('models/zenodo/bce/scan_bce_ECD_results.pkl', 'rb') as f:
    results = pickle.load(f)
    f.close()
odds_lrs = []
probit_lrs = []
arctan_lrs = []
# Filter results to have 10 initializations of each parametrization
for i, result in enumerate(results):
    optimizer_config = dict(
        optimizer = optimizer,
        learning_rate = result['learning_rate'],
        eta = result['eta'],
        F0 = result['F0'],
        nu = result['nu'],
    )
    classifier = result['classifier']
    if classifier == 'odds':
        model = create_model_original(d=6, loss_fun=bce, output='sigmoid', **optimizer_config)
        model.load_state_dict(torch.load(result['path'])['state_dict'])
        model.eval()
        odds_lrs.append(odds_lr(model, m, s))
    elif classifier == 'probit':
        model = create_model_original(d=6, loss_fun=probit_bce, output='linear', **optimizer_config)
        model.load_state_dict(torch.load(result['path'])['state_dict'])
        model.eval()
        probit_lrs.append(probit_lr(model, m, s))
    elif classifier == 'arctan':
        model = create_model_original(d=6, loss_fun=arctan_bce, output='linear', **optimizer_config)
        model.load_state_dict(torch.load(result['path'])['state_dict'])
        model.eval()
        arctan_lrs.append(arctan_lr(model, m, s))
```

```python
slice_index = 1
xs, plt_var = get_slice(slice_index)
lr_true = lr(torch.tensor(xs, dtype = torch.float)).detach().numpy()
lr_model = [odds_lrs[i](xs).detach().cpu().numpy() for i in range(len(odds_lrs))]
plt.plot(plt_var, lr_true, c='black', label='Truth')
(counts, bins) = np.histogram(X_mae[y_mae == 1][:,slice_index], bins = plt_var.reshape(-1)[::35], density=True)
plt.hist(bins[:-1], bins, weights = counts*50, alpha = 0.5, label = 'Monte Carlo')
(counts2, bins2) = np.histogram(X_mae[y_mae == 0][:,slice_index], bins = plt_var.reshape(-1)[::35], density=True)
plt.hist(bins2[:-1], bins2, weights = counts2*50, alpha = 0.5, label = 'data')
for i in range(10):
    plt.plot(plt_var, lr_model[i])
plt.xlim([plt_var[0], plt_var[-1]])
plt.ylim([0,4])
plt.xlabel('Multiplicity')
plt.ylabel('Likelihood Ratio')
plt.legend()
plt.savefig('NF_ECD_odds.png')
plt.show()
```
