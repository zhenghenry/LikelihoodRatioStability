# Obtaining all the metrics in the specificed problem and optimizer

```python
def obtain_sorted_dict(prob_type, optimizer, metric_to_sort, extension_name = ''):
    folder_path = os.path.join(prob_type, optimizer)
    metrics_file_list = glob.glob(folder_path + '/metrics/' + extension_name + '*.npy')
    print(folder_path)
    HP_sweep_dict = []
    for file_name in metrics_file_list:
        # Each file will contain a set of metrics we track for that problem and the optimizer config
        metrics = np.load(file_name, allow_pickle=True).item()
        if prob_type == 'ALEPH' or 'ALEPH_extend':
            n = metrics['n']
            n2_list = metrics['n2_preds']
            n2_list_new = []
            for elem in n2_list:
                n2_list_new.append(elem[:-9])
            n2_avg = metrics['n2_avg_pred'] 
            # print(n2_avg)
            weighted_mae = np.sum(n*np.mean(np.nan_to_num(np.abs(n2_list/n - 1)), axis = 0)/np.sum(n))
            weighted_error = np.sum(np.nan_to_num(n*np.abs(np.mean(n2_list/n-1, axis = 0)/np.sum(n))))
            weighted_mae_notail = np.sum(n[:-9]*np.mean(np.nan_to_num(np.abs(n2_list_new/n[:-9] - 1)), axis = 0)/np.sum(n[:-9]))
            weighted_error_notail = np.sum(np.abs(n[:-9]*np.nan_to_num(np.mean(n2_list_new/n[:-9]-1, axis = 0)/np.sum(n[:-9]))))
            n2_var_notail = np.var(n2_list, axis = 0)[:-9]
            weighted_var_notail = np.sum(n2_var_notail*n[:-9]/np.sum(n[:-9]))
            metrics.update(weighted_mae=weighted_mae)
            metrics.update(weighted_error=weighted_error)
            metrics.update(weighted_mae_notail=weighted_mae_notail)
            metrics.update(weighted_error_notail=weighted_error_notail)
            metrics.update(weighted_var_notail=weighted_var_notail)
        HP_sweep_dict.append(metrics)

    ### We can sort the list of dictionaries in ascending order according to 'weighted_mae' or any other scalar metric we choose.
    sorted_dict = sorted(HP_sweep_dict, key=lambda x: (np.round(x[metric_to_sort],4), x['weighted_var_notail']))
    ### Throw away metrics where the average prediction in the first bin contains nan
    sorted_dict = [element for element in sorted_dict if element['n2_avg_pred'][0] == element['n2_avg_pred'][0]]

    return sorted_dict
```

### Analyze ECD results

```python
fig, ax = plt.subplots(1,2, sharex=True, sharey=True, figsize=(15,8))
# fig.suptitle('Left: Individual Initializations | Right: Ensemble Average', fontsize=16)
ax[0].title.set_text('Individual Initializations')
ax[1].title.set_text('Ensemble Average')
for i, element in enumerate([ecd_dict[0]]):
    n = element['n']
    N_bin = N_tot*n/np.sum(n)
    n2_list = element['n2_preds']
    n2_avg = element['n2_avg_pred'] # *n/np.sum(n)
    n2_var = np.var(n2_list/n2_noweight, axis = 0)
    n2_std = np.std(n2_list/n2_noweight, axis = 0)
    n2_var_notail = n2_var[:-9]
    weighted_var_notail = np.sum(n2_var_notail*n[:-9]/np.sum(n[:-9]))
    # print(n2_avg)
    weighted_mae = element['weighted_mae_notail']
    weighted_error = element['weighted_error']
    for j in range(len(n2_list)):
        ax[0].plot(plot_points, n2_list[j]/n2_noweight)
    ax[1].errorbar(plot_points, n2_avg/n2_noweight, yerr=n2_std, c = 'black', label = f'Ensemble Average')
    eta = element['eta']
    print(f'MAE: {weighted_mae:.2e}, EA: {weighted_error:.2e}, eta: {eta}')
    ax[1].errorbar(plot_points, element['n']/element['n1'], yerr=error_bars, c='red', label = 'Truth', alpha = 0.3)
    ax[0].errorbar(plot_points, element['n']/element['n1'], yerr=error_bars, c='red', label = 'Truth', alpha = 0.3)
    ax[0].set_ylim([0.93,1.04])
    ax[1].set_ylim([0.93,1.04])
    ax[1].legend()
ax[0].set_xlim([0,0.35])
ax[0].set_xlabel('Thrust $1-T$', fontsize = 20)
ax[1].set_xlabel('Thrust $1-T$', fontsize = 20)
ax[0].set_ylabel('Likelihood Ratio')
plt.show()
```
